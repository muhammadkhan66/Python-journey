PART B - Whitespace and Cleaning

```python
_name = "        Abdul Rehman           "
print("Using rstrip: ", _name.rstrip())
print("using lstrip: ", _name.lstrip())
print("using strip: ", _name.strip())
```

```python
_string = "   Python Crash Course   "
print(_string.strip())
```
